```python
from bs4 import BeautifulSoup as bs
import requests as req
import re
import os
```

```python
folder_path = "project_gutenberg"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

url = "https://www.gutenberg.org/browse/languages/zh" 
res = req.get(url) 
soup = bs(res.text, "html.parser")

prefix = 'https://www.gutenberg.org'

chinese_pattern = re.compile('[\u4e00-\u9fff\r]')
id_pattern = r'/ebooks/(\d+)'

seen_titles = set()
count = 0

for li in soup.select('li.pgdbetext'):
    a = li.select_one('a')
    Title = a.get_text().replace('\r', '_').replace('\n', '_').replace('=', '_').replace('—', '_').replace(':', '_').strip()
    aLink = a['href']
    ebook_id = re.search(id_pattern,aLink).group(1)
    if chinese_pattern.search(Title) and Title not in seen_titles:
        if count <= 200 :
            count += 1
            suburl = f'https://www.gutenberg.org/cache/epub/{ebook_id}/pg{ebook_id}.txt'
            subres = req.get(suburl)
            matches_raw = re.findall(r'[\u4e00-\u9fff\uff00-\uffef\u3001-\u3015\n]', subres.text)
            text = re.sub(r'\n{3,}', '\n\n',"".join(matches_raw))
            with open(f"{folder_path}/{count}. {Title}.txt", "w", encoding="utf8") as file:
                file.write(text)
            seen_titles.add(Title)
            print(f'{count}. {Title}')
        else:
            break




```
